# Task 1 — Fetch and Validate API Response

```python
import requests

url = "https://restcountries.com/v3.1/all?fields=name,population,region,area"

try:
    response = requests.get(url)

    print("Status Code:", response.status_code)

    if response.status_code == 200:
        data = response.json()
        print("Total number of countries:", len(data))
    else:
        print("Error:", response.text)

except Exception as e:
    print("Exception occurred:", e)
```

# Task 2 — Extract Fields and Create DataFrame

```python
import pandas as pd

countries_list = []

for country in data:
    try:
        name = country["name"]["common"]
        population = country.get("population", 0)
        region = country.get("region", "Unknown")
        area = country.get("area", 0)

        countries_list.append({
            "common_name": name,
            "population": population,
            "region": region,
            "area": area
        })

    except KeyError:
        continue

df = pd.DataFrame(countries_list)

df.head(10)
```

# Task 3 — Data Summary

```python
# 1. Region with highest total population
region_population = df.groupby("region")["population"].sum()

highest_region = region_population.idxmax()
highest_population = region_population.max()

print("Region with Highest Total Population:")
print(f"{highest_region} → {highest_population}")


# 2. Country with largest area
largest_country = df.loc[df["area"].idxmax()]

print("\nCountry with Largest Area:")
print(f"{largest_country['common_name']} → {largest_country['area']}")
```
