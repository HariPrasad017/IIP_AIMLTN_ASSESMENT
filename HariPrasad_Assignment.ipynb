```python
import numpy as np

data = np.array([
    [22.5, 19.0, 31.2, 28.7, 25.1],
    [17.3, 22.8, 30.5, 26.4, 21.9],
    [33.1, 29.6, 18.4, 24.0, 27.8],
    [20.2, 23.5, 31.9, 28.1, 22.6]
])

print("Temperature Data:\n", data)


# -----------------------------
# Task 1
# -----------------------------
print("\nShape of data:", data.shape)

mean_per_station = np.mean(data, axis=1)

print("\nMean temperature per station:")
print(mean_per_station)


# -----------------------------
# Task 2
# -----------------------------

mask = data > 28.0

temps_above_28 = data[mask]

print("\nTemperature readings above 28.0°C:")
print(temps_above_28)


# -----------------------------
# Task 3
# -----------------------------

normalized = (data - data.min()) / (data.max() - data.min())

normalized_rounded = np.round(normalized, 2)

print("\nNormalized Temperature Data (0 to 1 range):")
print(normalized_rounded)
```
