# Introduction

Title:
E-commerce Sales Performance Analysis

Description

In this project, we analyze sales data from an e-commerce company to identify sales trends and top performers. Since the real database is unavailable, a synthetic dataset of 200 transactions is generated.

The analysis focuses on:

1. Sales performance by region

2. Transaction counts by product category

3. Performance comparison between salespersons

4. Multi-column grouping analysis

5. Custom aggregation functions

Using Pandas groupby operations, we generate summary reports to support management decision-making.

# Import Libraries

```python
import pandas as pd
import random
```

Explanation

1. Pandas is used for data manipulation and analysis.

2. Random is used to generate synthetic sales data.

# Generate Synthetic Dataset

```python
random.seed(42)

regions = ['North', 'South', 'East', 'West']
categories = ['Electronics', 'Clothing', 'Home & Garden', 'Sports', 'Books']
salespersons = ['Alice', 'Bob', 'Carol', 'David', 'Emma', 'Frank']

data = {
    'transaction_id': range(1001, 1201),
    'region': [random.choice(regions) for _ in range(200)],
    'category': [random.choice(categories) for _ in range(200)],
    'salesperson': [random.choice(salespersons) for _ in range(200)],
    'sales_amount': [round(random.uniform(50, 5000), 2) for _ in range(200)],
    'customer_id': [random.randint(5000, 5100) for _ in range(200)]
}

df = pd.DataFrame(data)

print(df.head(10))
print("\nDataset shape:", df.shape)
```

Explanation

This code:

1. Defines possible values for:



*   Regions
*   Product Categories


*   Salespersons


2. Generates 200 random transactions

3. Creates a Pandas DataFrame

| Column         | Description            |
| -------------- | ---------------------- |
| transaction_id | unique transaction     |
| region         | sales region           |
| category       | product category       |
| salesperson    | employee who made sale |
| sales_amount   | transaction value      |
| customer_id    | customer identifier    |


# TASK 1: Basic Grouping and Aggregations

**Total Sales by** **Region**

```python
regional_sales = df.groupby('region')['sales_amount'].sum().reset_index()

regional_sales = regional_sales.sort_values(by='sales_amount', ascending=False)

print("Total Sales by Region")
display(regional_sales)
```

Explanation

groupby('region') groups data by region.

sales_amount.sum() calculates total sales per region.

reset_index() converts the grouped result into a clean DataFrame.

sort_values() sorts regions from highest to lowest sales.

This helps identify the top-performing region.

# Transaction Count by Category

```python
category_transactions = df.groupby('category')['transaction_id'].count().reset_index()

print("Transactions per Category")
display(category_transactions)
```

Explanation

This calculates number of transactions per product category.

We use:

groupby('category')
count()

This shows which product categories are most popular.

# Average Sales per Salesperson

```python
avg_salesperson_sales = df.groupby('salesperson')['sales_amount'].mean().reset_index()

print("Average Sales per Salesperson")
display(avg_salesperson_sales)
```

Explanation

This calculates the average transaction value for each salesperson.

mean() computes the average sales amount.

This helps evaluate salesperson performance.

# TASK 2: Multi-Column Grouping

**Sales by Region AND Category**

```python
region_category_sales = df.groupby(['region', 'category'])['sales_amount'].sum().reset_index()

print("Sales by Region and Category")
display(region_category_sales)
```

Explanation

This groups data using two columns simultaneously:

region
category

It shows which product categories perform best in each region.

Example insight:

1. Electronics may sell more in North

2. Clothing may sell more in South

# Multiple Aggregations for Salespersons

```python
salesperson_performance = df.groupby('salesperson')['sales_amount'].agg(['sum', 'mean', 'count']).reset_index()

print("Salesperson Performance")
display(salesperson_performance)
```

Explanation

The .agg() function allows multiple metrics at once.

Metrics calculated:

# Sort Salespersons by Total Sales

```python
salesperson_performance = salesperson_performance.sort_values(by='sum', ascending=False)

print("Top Performing Salespersons")
display(salesperson_performance)
```

Explanation

Sorting by sum identifies the top sales performer.

# Find Category with Maximum Revenue

```python
category_sales = df.groupby('category')['sales_amount'].sum()

top_category = category_sales.idxmax()

print("Category with Maximum Revenue:", top_category)
```

Explanation

idxmax() finds the index with highest value.

So it returns the category generating the most revenue.

# TASK 3: Custom Aggregation

**Define Custom** **Function**

```python
def sales_range(x):
    return x.max() - x.min()
```

Explanation

This function calculates:

 sales range = maximum sale - minimum sale

It measures variation in sales amounts.

# Apply Custom Aggregation

```python
region_analysis = df.groupby('region')['sales_amount'].agg(['sum','mean','min','max', sales_range]).reset_index()

print("Region Sales Analysis")
display(region_analysis)
```

# Final Sales Summary Report

```python
final_report = df.groupby('region').agg({
    'sales_amount': ['sum', 'mean'],
    'customer_id': 'count'
}).reset_index()

print("Final Sales Summary Report")
display(final_report)
```

# Multi-Level Column Structure Explanation

When we use dictionary-based aggregation like:

df.groupby('region').agg({

'sales_amount':['sum','mean'],

'customer_id':'count'

})

Pandas creates multi-level column names.

| region | sales_amount | sales_amount | customer_id |
| ------ | ------------ | ------------ | ----------- |
|        | sum          | mean         | count       |

This happens because:

1. One column (sales_amount) has multiple aggregations

2. Pandas organizes them in hierarchical columns

This differs from single aggregation because:

Single aggregation produces one column only, while multi aggregation produces multiple statistical measures per column.
