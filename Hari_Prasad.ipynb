# Import Libraries

```python
import pandas as pd
import numpy as np
```

Explanation

Pandas is used for data manipulation and analysis. NumPy is used for numerical operations.

# Task 1 – Data Preparation

**Step 1: Create** **DataFrames**

```python
students_data = {
    'student_id': [101, 102, 103, 104, 105, 106, 107],
    'name': ['Alice', 'Bob', None, 'David', 'Emma', 'Frank', 'Grace'],
    'email': ['alice@email.com', 'bob@email.com', 'charlie@email.com', None, 'emma@email.com', 'frank@email.com', 'grace@email.com'],
    'city': ['Mumbai', 'Delhi', 'Bangalore', 'Mumbai', None, 'Chennai', 'Delhi']
}

enrollments_data = {
    'student_id': [101, 102, 103, 105, 108, 109],
    'course_name': ['Python', 'Data Science', 'Python', 'Machine Learning', 'AI', 'Python'],
    'enrollment_date': ['2024-01-15', '2024-01-20', '2024-02-01', '2024-02-10', '2024-02-15', '2024-03-01']
}

scores_data = {
    'student_id': [101, 102, 104, 105, 106],
    'exam_score': [85, 92, 78, 88, 95]
}

students = pd.DataFrame(students_data)
enrollments = pd.DataFrame(enrollments_data)
scores = pd.DataFrame(scores_data)

print("Original Students DataFrame")
display(students)
```

Explanation

This step converts dictionary data into Pandas DataFrames which makes the data easier to manipulate.

**Step 2: Null Value** **Analysis**

```python
print("Null Value Analysis")

null_counts = students.isnull().sum()
null_percentage = (students.isnull().sum() / len(students)) * 100

for col in students.columns:
    print(f"Column: {col}, Nulls: {null_counts[col]} ({null_percentage[col]:.2f}%)")
```

Explanation:

1. isnull() detects missing values.

2. sum() counts them.

3. Percentage helps understand how much data is missing.

# Step 3: Fill Missing City Values

```python
students['city'] = students['city'].fillna('Unknown')
```

Explanation:

Missing values in city are replaced with "Unknown".

# Step 4: Drop Missing Names

```python
students_clean = students.dropna(subset=['name'])

print("Cleaned Students DataFrame")
display(students_clean)
```

Explanation:

1. Rows with missing names are removed.

2. Because student identity cannot be missing.

Final rows = 6 students

# Task 2 – Join Operations

**Inner** **Join**

```python
inner_join = pd.merge(students_clean, enrollments, on='student_id', how='inner')

print("Inner Join Result")
display(inner_join)

print("Total rows:", len(inner_join))
```

Explanation:

Inner join keeps only matching student_ids in both tables.

Result students:

101 Alice

102 Bob

105 Emma

Excluded students:

104, 106, 107

Reason:

They are not present in enrollments table.

# Left Join

```python
left_join = pd.merge(students_clean, enrollments, on='student_id', how='left')

print("Left Join Result")
display(left_join)

print("Total rows:", len(left_join))
```

Explanation:

Left join keeps all students, even if they are not enrolled.

Students with null course_name:

104, 106, 107

Reason:

They did not enroll in any course.

```python
right_join = pd.merge(students_clean, enrollments, on='student_id', how='right')

print("Right Join Result")
display(right_join)

print("Total rows:", len(right_join))
```

Explanation:

Right join keeps all enrollments.

Student IDs without names:

108, 109

Reason:

These IDs exist in enrollments but not in students dataset.

# Full Outer Join

```python
outer_join = pd.merge(students_clean, enrollments, on='student_id', how='outer')

print("Outer Join Result")
display(outer_join)

print("Total rows:", len(outer_join))
```

Explanation:

Outer join keeps all rows from both tables.

# Rows with Missing Data

```python
missing_rows = outer_join[(outer_join['name'].isnull()) | (outer_join['course_name'].isnull())]

display(missing_rows)
```

Explanation:

This shows students without courses OR enrollments without student data.

# Merge Indicator

```python
outer_indicator = pd.merge(students_clean, enrollments, on='student_id', how='outer', indicator=True)

display(outer_indicator)

print("Merge Source Distribution")
print(outer_indicator['_merge'].value_counts())
```

# Task 3 – Lookup Operation

**Dictionary** **Mapping**

```python
score_dict = scores.set_index('student_id')['exam_score'].to_dict()

print(score_dict)
```

Explanation:

Creates dictionary like:

{101:85,102:92,104:78,105:88,106:95}

**Add Scores using map()**

```python
students_clean['exam_score'] = students_clean['student_id'].map(score_dict)

display(students_clean)
```

Explanation:

.map() performs fast lookup using dictionary keys.

Students without scores get NaN.

# Merge Method

```python
merge_scores = pd.merge(students_clean, scores, on='student_id', how='left')

display(merge_scores)
```

Explanation:

This performs the same operation using merge.

# Why map() is Faster

Write this in Markdown:

Lookup using .map() is faster than merge() when we only need to map one column because:

1. It uses dictionary-based lookup (O(1) time complexity)

2. It avoids creating a full join operation

3. It uses less memory

4. It is simpler when mapping one key to one value

Merge is better when combining multiple columns or complex datasets.

# Automation Function

```python
def auto_merge(df1, df2, join_type, key_column):

    merged_df = pd.merge(df1, df2, on=key_column, how=join_type)

    result = {
        'result_df': merged_df,
        'row_count': len(merged_df),
        'join_type': join_type
    }

    return result
```

Explanation:

This function:

1. Accepts two dataframes

2. Accepts join type

3. Performs merge

4. Returns dictionary with results

# Test the Function

```python
test_inner = auto_merge(students_clean, enrollments, 'inner', 'student_id')

print("Join Type:", test_inner['join_type'])
print("Rows:", test_inner['row_count'])

display(test_inner['result_df'])
```

**Second Test**

```python
test_left = auto_merge(students_clean, enrollments, 'left', 'student_id')

print("Join Type:", test_left['join_type'])
print("Rows:", test_left['row_count'])

display(test_left['result_df'])
```
