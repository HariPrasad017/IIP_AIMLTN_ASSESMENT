This step imports the pandas library which is used for data manipulation and analysis.

```python
import pandas as pd
```

This step creates a pandas DataFrame using the provided student dataset.

```python
data = {
    "StudentID": [101,102,103,104,105,106,107,108,109,110],
    "Name": ["Alice","Bob","Charlie","Diana","Ethan","Fiona","George","Hannah","Ivan","Julia"],
    "Age": [20,21,19,22,20,21,23,19,22,20],
    "Marks": [88,75,92,65,78,85,70,95,60,80],
    "Grade": ["A","B","A","C","B","A","B","A","C","B"]
}

df = pd.DataFrame(data)

print("Student DataFrame:")
print(df)
```

This step saves the DataFrame into a CSV file named students.csv.

```python
df.to_csv("students.csv", index=False)

print("students.csv file saved successfully.")
```

This step loads the saved students.csv file into a new DataFrame.

```python
df_new = pd.read_csv("students.csv")

print("Reloaded DataFrame:")
print(df_new)
```

This step displays the first 3 rows of the DataFrame using head().

```python
print("First 3 rows of the DataFrame:")
print(df_new.head(3))
```

This step shows the statistical summary of numerical columns using describe().

```python
print("Statistical Summary:")
print(df_new.describe())
```

This step displays the number of rows and columns in the DataFrame.

```python
print("Shape of the DataFrame:")
print(df_new.shape)
```

This step shows the data types of each column in the DataFrame.

```python
print("Data types of each column:")
print(df_new.dtypes)
```

This step displays detailed information about the DataFrame including memory usage and non-null counts.

```python
print("Detailed DataFrame Information:")
df_new.info()
```
