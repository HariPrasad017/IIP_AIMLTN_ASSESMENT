```
print("===== Task 1: Find Maximum and Minimum =====")

temperatures = [28, 32, 35, 29, 31, 27, 30]







if not temperatures:

  print ("The list is empty.")

else:

  max_temp = temperatures[0]

  min_temp = temperatures[0]







for temp in temperatures:

  if temp > max_temp:

    max_temp = temp

  if temp < min_temp:

    min_temp = temp







print (f"Highest Temperature: {max_temp} °C")

print (f"Lowest Temperature: {min_temp} °C")







print("\n===== Task 2: Count Hot Days ====")



temperatures = [28, 32, 35, 29, 31, 27, 30]

hot_days_count = 0







for temp in temperatures:

  if temp <= 30:

    continue

  hot_days_count += 1

print(f"Hoy Days (>30℃): {hot_days_count}")









print("\n===== Task 3: Alert System ====")



temperatures = [28, 32, 35, 40, 31, 33, 30]







hot_days_count = 0

day_counter = 0

for temp in temperatures:

  day_counter +=1

  if temp >=40:

    print (f"Alert! Extreme Temperature {temp} ℃ detected on Day {day_counter})")

    break

  if temp <=30:

    continue

  hot_days_count += 1

print(f"Hot Days before alert:{hot_days_count}")
```
