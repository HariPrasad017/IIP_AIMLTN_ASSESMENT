# Import Libraries

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

# Load dataset

```python
tips = sns.load_dataset("tips")
```

# Preview data

```python
tips.head()
```

# Task 1 — Matplotlib Bar Chart

# Prepare Data

```python
tips_per_day = tips.groupby("day")["tip"].sum()

tips_per_day
```

# Create Bar Chart

```python
plt.figure(figsize=(8, 5))

plt.bar(tips_per_day.index, tips_per_day.values , color = "violet")

plt.title("Total Tips Collected per Day")
plt.xlabel("Day")
plt.ylabel("Total Tip Amount")

plt.show()
```

# Seaborn Histogram

# Create Histogram

```python
plt.figure(figsize=(8, 5))

sns.histplot(data=tips, x="tip", bins=20, kde=True,color="black")

plt.title("Distribution of Tip Amounts")

plt.show()
```

The histogram shows that most tip amounts fall between 2 and 4 dollars, indicating that customers usually give moderate tips.

The distribution is slightly right-skewed, meaning there are fewer large tips compared to smaller ones. The KDE curve helps smooth the distribution and makes it easier to observe the overall pattern.

# Task 3 — Seaborn Scatter Plot

# Create Scatter Plot

```python
plt.figure(figsize=(8, 5))

sns.scatterplot(data=tips, x="total_bill", y="tip", hue="day")

plt.title("Total Bill vs Tip Amount")

plt.show()
```

The scatter plot shows a positive relationship between total bill and tip amount, meaning higher bills generally result in higher tips.

Different colors indicate that this trend is consistent across different days of the week.
