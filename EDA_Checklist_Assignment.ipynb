# STEP 2: Generate Dataset

```python
import pandas as pd
import numpy as np

np.random.seed(42)

n = 200

data = {
    "order_id": range(1001, 1001 + n),
    "city": np.random.choice(["Mumbai", "Delhi", "Bangalore", "Chennai"], size=n),
    "category": np.random.choice(["Electronics", "Clothing", "Groceries", "Furniture"], size=n),
    "order_value": np.random.randint(200, 5000, size=n).astype(float),
    "delivery_days": np.random.randint(1, 15, size=n).astype(float),
    "rating": np.random.choice([1, 2, 3, 4, 5, None], size=n)
}

missing_indices_order = np.random.choice(n, size=15, replace=False)
missing_indices_delivery = np.random.choice(n, size=10, replace=False)
data["order_value"][missing_indices_order] = np.nan
data["delivery_days"][missing_indices_delivery] = np.nan

data["order_value"][5] = 95000
data["order_value"][88] = 87000

df = pd.DataFrame(data)

print(df.shape)
df.head()
```

# Task 1 — Missing Values Analysis

**Shape**

```python
print("Shape of dataset:", df.shape)
```

**Data Types**

```python
df.dtypes
```

**Missing Values Count**

```python
df.isnull().sum()
```

**Missing Percentage**

```python
missing_percent = (df.isnull().sum() / len(df)) * 100
missing_percent.round(2)
```

# Task 2 — Summary & Visualization

```python
df.describe()
```

1. The order_value column has a very high maximum value compared to its median, indicating the presence of outliers.

2. The delivery_days column has a relatively small range, showing that most deliveries happen within a consistent number of days.

**Histogram (order_value)**

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.hist(df["order_value"].dropna(), color="Brown")

plt.title("Distribution of Order Value")
plt.xlabel("Order Value")
plt.ylabel("Frequency")

plt.show()
```

**Box Plot (delivery_days)**

```python
plt.figure(figsize=(8, 5))

plt.boxplot(df["delivery_days"].dropna())

plt.title("Box Plot of Delivery Days")
plt.ylabel("Delivery Days")

plt.show()
```

# Task 3 — Correlation Analysis

**Correlation Matrix**

```python
corr_matrix = df.corr(numeric_only=True)
corr_matrix
```

**Heatmap**

```python
import seaborn as sns

plt.figure(figsize=(8, 5))

sns.heatmap(corr_matrix, annot=True)

plt.title("Correlation Heatmap")

plt.show()
```

There is a weak positive relationship between order_value and delivery_days, suggesting that higher-value orders may take slightly longer to deliver.
